# Центральная пределеньная теорема своими руками

```python
import numpy as np
import math
import scipy.stats as sts
from scipy.stats import expon
import matplotlib.pyplot as plt
%matplotlib inline
```

## Экспоненциальное распределение

Для выполнения задания, мною было выбрано [экспоненциальное распределение](https://ru.wikipedia.org/wiki/%D0%AD%D0%BA%D1%81%D0%BF%D0%BE%D0%BD%D0%B5%D0%BD%D1%86%D0%B8%D0%B0%D0%BB%D1%8C%D0%BD%D0%BE%D0%B5_%D1%80%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5).

Функция экспоненциального распределения:
$$F_{X}(x)=\begin{cases}1-e^{{-\lambda x}}&,\;x\geq 0\\0&,\;x<0\end{cases}$$

```python
#генерируем выборку объемом 1000
exp=expon() #создаем функцию
y=exp.rvs(size=1000) #генерируем случайную выборку
```

```python
x = np.linspace(0,10,1000) #определяем массив точек для построения плотности распределения
pdf=exp.pdf(x)#вычисляем значения плотности распределения от выборки
```

```python
#строим гистрограмму выборки и теоретическую функцию распределения
plt.figure(figsize=(8,5))
plt.plot(x, pdf, label='Теоретическое распределение')
plt.hist(y,bins=50, range=(0,8), normed=True, label="Значения")
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='best')
```

Среднее:
$${\mathbb  {E}}[X]={\frac  {1}{\lambda }}$$

Дисперсия:
$$\operatorname {D}[X]={\frac  {1}{\lambda ^{2}}}$$, где $$\lambda >0$$

```python
#считаем среднее  и диспресию с помощью создания функций
def mean(l):
    return (1/l) #где l - lambda
def disp(l):
    return (1/(l**2))
```

```python
l=1 #из графика мы видим, что в функции была использовано значение lambda=1
E=mean(l) #считаем среднее
D=disp(l) #считаем дисперсию
print ("E(X) =", E)
print ("D(X) =", D)
```

## Оценка распределения выборочного среднего случайной величины при разных объёмах выборок

### Объем выборки n=3

```python
n=3
y_3=np.array([]) #обозначим массив выборочных средних
i=1
while i <= 1000: #цикл генерации случайных выборок количества n из функции
    y_3i = exp.rvs(n) #генерация выборки
    E_3i=sum(y_3i)/n #находим выборочное среднее
    y_3=np.append(y_3,E_3i) #добавляем в массив полученное значение выборочного среднего
    i+=1
D_3=D/n #дисперсия нормального распределения выборки средних
Sq_3=math.sqrt(D_3) #стандартное отклонение нормального распределения 
norm_3 = sts.norm(E, Sq_3) #создаем нормальное распределение с теорет. средним  expon и со станд.отклонением выборки средних
pdf_3 = norm_3.pdf(x[:500]) #определяем плотность распределения
#строим гистрограмму выборочных средних при т=3 и плотность полученного распределения
plt.figure(figsize=(8,5))
plt.plot(x[:500], pdf_3, label='PDF(n=3)')
plt.hist(y_3,bins=50, range=(0,5), normed=True, label='Sample')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='best')
```
